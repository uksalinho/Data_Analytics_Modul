## K-Nearest Neighbors(KNN)

```python
# a lazy learner : eğitim yok
# data daki değişkenleri 2 boyutlu koordinatlarına map liyor, hafızasına alıyor
# yeni yapılan gözlem çevresindeki class lara göre label lanıyor

# non-linear non-parametric : katsayı yok

# küçük datasetlerinde daha fazla kullanılıyor. 
# false-detection modellerde, recommender system modellerde sıklıkla kullanılır
```

```python
# eğer k küçük olursa model OVERFITTING olur. 
# 1 komşuya göre modelleme yaptığımızda her noktaya uğramaya temas etmeye çalışıyor
# variance artıyor
# k arttığında ise model UNDERFITTING e gidebiliyor. Ayrıca k sayısını artırmak maliyeti de artırıyor
```

```python
# yeni bir veri geldiğinde 
# yeni gözlemi datadaki tüm noktalara mesafesini alıyor
# hesapladığı bu mesafeleri sıralıyor
# buna göre belirlenen k sayısına uygun olarak en yakın noktaları yapılan sıralamaya göre belirliyor

# büyük datasetleri için bunu her seferinde yapıyor. dolayısı ile maliyet büyük oluyor. küçük datasetlerinde kullanımı daha uygun
```

```python
# hatanın en düşük olduğu nokta k sayısının 35 civarında olduğu sağdaki bölüm
# bu noktalarda modelin komplexity si çok fazla
# ayrıca UNDERFITTING tehlikesi de çok artıyor
# k sayısı gözlem saysına ne kadar yaklaşırsa prediction lar majority class lehine değişecektir.

# dolayısı ile k sayısı için bir denge noktası bulmamız lazım
# maliyet-loss-score durumuna göre k yı olabildiğince en küçük olduğu yerden seçmek lazım

# gridsearch ile k tespit edilebilir. ama gridsearch min loss a göre k belirleyeceğinden UNDERFITTING tehlikesi artar
# bu nedenle ELBOW kullanılması tavsiye edilir
```

```python
# UNIFORM
# yeni gözlemin etrafındaki k=4 e göre en yakın 4 noktadan ağırlığı en fazla olana göre sınıflandırma yapıyor
```

```python
# DISTANCE
# mesafeye göre bir ağırlıklandırma yapıyor
# dolayısı ile k=4 e göre en yakın 4 noktayı hem mesafe hem de ağırlıklarına göre değerlendiriyor ve buna göre sınıflanduırma yapıyor
```

```python
# MATHATTAN outlier bakımından EUCLIDEAN a göre daha hassas
```

```python
# Yeni gözlemin UZUN aralıkta olan feature lara atanma ihtimali YAKIN aralıkta olan feature lara göre daha az oluyor
# bu nedenle SCALE mutlaka yapılmalı
# KNN algoritmasının DISTANCE tabanlı olması SCALE yapılmasını zaruri hale getiriyor.
```

```python
# corelasyon linearity katsayı gibi herhangi bir assumption aramıyor.
# feature ların birbirine çok girift olduğu datalarada iyi sonuç vermiyor
# küçük data setleri için daha uygun
# yüksek boyutlu veri setlerind eiyi çalışmıyor
```

## Elbow Method for Choosing Reasonable K Values

```python
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score
```

```python
test_error_rates = []


for k in range(1,30):   # k nin cok büyük olmasi  tercih edilmez. hesaplama maliyetinden dolayi 
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(X_train_scaled,y_train) 
   
    y_pred_test = knn_model.predict(X_test_scaled)
    
    
    test_error = 1 - accuracy_score(y_test,y_pred_test) # accuracy alınmasının nedeni data nın dengeli olması
                                                        # modelin ve datanın durumuna göre değişebilir
    
    test_error_rates.append(test_error)
```

```python
# error rates isimli boş bir list oluşturuyoruz
# k için bir aralık belirliyoruz
# belirlediğimiz k aralığındaki her bir k değeri için hata hesaplanıp boş listeye ekliyoruz.
```

```python
# burasi yukari satirin görsellestirilmesi

plt.figure(figsize=(15,8))
plt.plot(range(1,30), test_error_rates, color='blue', linestyle='--', marker='o',
         markerfacecolor='red', markersize=10)

plt.title('Error Rate vs. K Value')
plt.xlabel('K_values')
plt.ylabel('Error Rate')
plt.hlines(y=0.050, xmin = 0, xmax = 30, colors= 'r', linestyles="--")
plt.hlines(y=0.057, xmin = 0, xmax = 30, colors= 'r', linestyles="--")
plt.show()

# y'nin minimum olduğu iki değer için 0.05 ve 0.06) çizgi çizdirdik.

# x değeri olarak 1 den 30 kadar range yaptım. y olarak da test error rates leri atadım.

# amacım minimum error skorunu veren k sayısını bulmak.

# genel amacımız minimum complexcity ile maksimum verimi almaktır. o yüzden k sayısını optimum olarak belirlemeliyim.
```

```python
# üstteki kirmiziya göre error = 0.06
# alttaki kirmiziya göre error = 0.05, k büyüyor hesaplama maliyeti artiyor. Underffititng olma olasilgi artiyor.
```

```python
# x-ekseni: yukarıda belirlediğimiz k aralığı
# y-ekseni: yukarıda aldığımız hata değerleri
# 
```

```python
# k =1 iken ucuruum var. ideal nokta 5 ile 10 arasi kirmizi nokta min oldugu yer k=9 
# k = 23 de error min noktasi ama hesaplama maliyeti var. 0.01 error degeri icin degmez.
```

## Overfiting and underfiting control for k values

```python
test_error_rates = []
train_error_rates = []


for k in range(1,30):
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(X_train_scaled,y_train) 
   
    y_pred_test = knn_model.predict(X_test_scaled)
    y_pred_train = knn_model.predict(X_train_scaled)
    
    test_error = 1 - accuracy_score(y_test,y_pred_test)
    train_error = 1 - accuracy_score(y_train,y_pred_train)
    test_error_rates.append(test_error)
    train_error_rates.append(train_error)
```

```python
plt.figure(figsize=(15,8))
plt.plot(range(1,30), test_error_rates, color='blue', linestyle='--', marker='o',
         markerfacecolor='red', markersize=10)
plt.plot(range(1,30), train_error_rates, color='green', linestyle='--', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K_values')
plt.ylabel('Error Rate')
plt.hlines(y=0.050, xmin = 0, xmax = 30, colors= 'r', linestyles="--")
plt.hlines(y=0.057, xmin = 0, xmax = 30, colors= 'r', linestyles="--")
plt.show()
```

```python
# MAVİ ---> TRAIN SETİ
# KIRMIZI ---> TEST SETİ
```

```python
# sağda % 1 bir iyileşme sağlıyor. bununla beraber k sayısı arttığından hesaplama maliyeti ve UNDERFITTING tehlikesi artıyor.
# solda ise hata oranı %1 gibi daha fazla
# dolayısıyla bir TRADE OFF durumu söz konusu
# makineleriniz çok iyi ise maliyet sıkıntısı yok ise k sayısı için sağdaki tercih uygun olabilir
```

## Scores by Various K Values

```python
def eval_metric(model, X_train, y_train, X_test, y_test):
    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    
    print("Test_Set\n")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print()
    print("Train_Set\n")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
```

```python
# FIRST A QUICK COMPARISON TO OUR ORIGINAL K=1

knn = KNeighborsClassifier(n_neighbors=1)

knn.fit(X_train_scaled,y_train)
print('WITH K=1\n')
eval_metric(knn, X_train_scaled, y_train, X_test_scaled, y_test)
```

```python
# NOW WITH K=22

knn = KNeighborsClassifier(n_neighbors=22)

knn.fit(X_train_scaled,y_train)
print('WITH K=22\n')
eval_metric(knn, X_train_scaled, y_train, X_test_scaled, y_test)
```

```python
# NOW WITH K=9

knn = KNeighborsClassifier(n_neighbors=9)

knn.fit(X_train_scaled,y_train)
print('WITH K=9\n')
eval_metric(knn, X_train_scaled, y_train, X_test_scaled, y_test)
```

```python
# NOW WITH K=15

knn = KNeighborsClassifier(n_neighbors=15)

knn.fit(X_train_scaled,y_train)
print('WITH K=15\n')
eval_metric(knn, X_train_scaled, y_train, X_test_scaled, y_test)
```

## Pipeline

What happens can be described as follows:

Step 1: The data are split into TRAINING data and TEST data according to ratio of train_test_split

Step 2: the scaler is fitted on the TRAINING data

Step 3: the scaler transforms TRAINING data

Step 4: the models are fitted/trained using the transformed TRAINING data

Step 5: the scaler is used to transform the TEST data

Step 6: the trained models predict using the transformed TEST data

pipe.fit(X_train, y_train)--> scaler.fit_transform(X_train) --> knn.fit(scaled_X_train, y_train)

pipe.predict(X_test)      --> scaler.transform(X_test)      --> knn.predict(scaled_X_test)

```python
# pipeline ile yapılan işlem;

# FIT-TRANSFORM ve FIT-PREDICT işlemlerini yapan algoritmaları OPTİMİZE ediyor
```

```python
from sklearn.pipeline import Pipeline
```

```python
operations = [("scaler", StandardScaler()), ("knn", KNeighborsClassifier())]

# burada diyorum ki; data mı önce StandardScaler ile scale et, ardından KNeighboursClassifier uygula
# operations içine yazdığım ilk algoritmalar FIT-TRANSFORM algoritmaları
# operations içine yazdığım en sonda FIT-PREDICT yapan ML algoritmaları olacak

# bütün algoritmalar icin kullanilabilir.
# ilk tupple a scale yazilmali(fit-transform yapacak), ikinci tupple(fit-prediction yapacak) a model i yazdik
```

```python
Pipeline(steps=operations)

# operations ile belirlenen stepleri benim yerime yap.
```

```python
pipe_model = Pipeline(steps=operations)
```

```python
pipe_model.fit(X_train, y_train)

# artık pipe_model.fit dediğim zaman benim için şunları yapıyor:
# scaler.fit_transform(X_train) --> knn.fit(scaled_X_train, y_train)

# yaptigi islem; scaler.fit_transform(X_train) --> knn.fit(scaled_X_train, y_train) (scaled edilmis haliyle knn de fit yapiyo)
# pipe.predict(X_test) --> scaler.transform(X_test) --> knn.predict(scaled_X_test)
```

```python
# Scale edilmemis degerler giriliyor, zaten kendisi scale yapiyor

y_pred = pipe_model.predict(X_test)
y_pred

# pipe_model.predict(X_test) benim için şunları yapıyor --> scaler.transform(X_test) --> knn.predict(scaled_X_test)
# predict edeceğim datayı (X_test) önce scale etmem gerekiyordu (modelimi train ederken X_train datamı scale ettiğim için)
# Pipeline, bu işlemleri sırasıyla yapıyor
```

## CrossValidate on Pipeline

```python
# CROSS VALIDATION ı buraya kadar hep SCALED TRAIN SET ile yaptık

# SCALED TRAIN SET inin tamamı tüm train setinin MEAN ve STD bilgilerini biliyor

# Dolayısıyla cv=5 belirlediğimde 4 parça TRAIN ve 1 parça VALIDATION oluyor
# TRAIN SET VALIDATION SET deki, MEAN ve STD bilgilerini bildiğinden doğal olarak TRAIN SET indeki tüm gözlemlerin bilgilerine sahip
# Bu yüzden CV yapılırken DATA LEAKAGE oluyor.


# BUnu önlemek için SCALE edilmiş her data da CV yaparken bundan sonra DATA LEAKAGE önlemek için PIPELINE kullanacağız.
# PIPELINE ne yapacak?

# PIPELINE da SCALE EDİLMEMİŞ TRAIN SET i kullanıyoruz.

# CV sayısı kadar cross validation yapacak
# Ancak, burada her cv de model sıfırlanacağından TRAIN SET i VALIDATION SET deki MEAN ve STD bilgilerini bilemeyecek
# Scale yapılırken X_TRAIN deki tüm veriler birbirini görüyor. Sıkıntı bu.

# Ve böylece DATA LEAKAGE önlenmiş olacak
```

```python
operations = [('scaler',StandardScaler()),('knn',KNeighborsClassifier(n_neighbors=9))]
model = Pipeline(operations)

scores = cross_validate(model, X_train, y_train, scoring = ['precision','recall','f1','accuracy'], cv = 10)
df_scores = pd.DataFrame(scores, index = range(1, 11))
df_scores.mean()[2:]

# Data Leakage olusmamasi icin scale edilmemis train datasi ile pipeline CV yapilmasi gerekir.
```
