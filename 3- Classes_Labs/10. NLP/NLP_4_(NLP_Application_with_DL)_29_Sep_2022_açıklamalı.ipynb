- LSTM'de 3 kap var:
- Forget gate = uzun sreli hafzmda tutlmamas gerek bilgileri silmeye yarar
- LSTM mimarisinde temsil edilen ilk blok, unut kapsdr (ft). Mevcut giriten (Xt) ve nceki gizli durumdan (ht) gelen bilgiler sigmoid aktivasyon fonksiyonundan geirilir. k deeri 0'a yaknsa unutma, 1'e yakn ise tutma anlamna gelir.
- Input gate = uzun sreli hafzamda tutulmas istenen ve gncellenenbilgiler tutulur.
- Hcre durumuna bir girdi olarak alr. ki blmden oluur; ilk olarak, hangi deerlerin gncelleneceine karar vermek iin nceki gizli durumu (ht) ve mevcut girii (Xt) bir sigmoid fonksiyonuna geiriyoruz. Ardndan, a dzenlemek iin ayn iki girii tanh aktivasyonuna geirin. Son olarak, hcre durumunu gncellemek iin hangi bilgilerin nemli olduuna karar vermek iin tanh ktsn (C't) sigmoid kts (it) ile arpn.
- Output gate = cel state ile hidden state i birletirir.
- Gizli durum, nceki girdilerle ilgili bilgileri ierir ve tahmin iin kullanlr. k kaps, mevcut gizli durumu (ht) dzenler. nceki gizli durum (ht-1) ve mevcut giri (xt), sigmoid ilevine iletilir. Bu kt, mevcut gizli durumu elde etmek iin tanh fonksiyonunun kts ile arplr. Mevcut durum (Ct) ve mevcut gizli durum (ht), klasik bir LSTM nitesinin son ktlardr.
- Cell State = nceki hcre durumundan (Ct-1) gelen girdi, unutma kaps ktsyla noktasal olarak arplr. Unutma kts 0 ise, nceki hcre ktsn (Ct-1) atacaktr. Bu k, yeni hcre durumunu (Ct) gncellemek iin giri kaps kna noktasal olarak eklenir. Mevcut hcre durumu, bir sonraki LSTM biriminin girii olacaktr.
- Input gate, LSTM biriminin mevcut hcre durumunda hangi bilgilerin gncellenmesiyle ilgili olduuna karar verir. Output gate, bir sonraki LSTM birimine geirilecek olan mevcut gizli durumu belirler.

```python
# LSTM den daha hzl alr. (GRU)
# update gate: Forget gate ve input gate in karm gibi neyi unutup neyi tutacam.
# reset gate: ne kadarn aklmda tutacam ayarlar.
# lstm i ksaltm. 2 gate e drm. lemi hzlandrm.
```

## Modelling

```python
model = Sequential()

# modelimi sequential ile balatyorum.
```

```python
embedding_size = 50

# word embedding de EMBEDING LAYER var
# benim modelime verdiim o dizinlerin ,ierisinde hatrlarsanz her birinde 61 token olacak. 
# O tokenlerin her birini 50 lik vektrlere evir diyorum.
```

```python
# pip install -U numpy==1.19.5
```

```python
model.add(Embedding(input_dim=num_words, # ilk layer mutlaka kelime vektrlerinden oluan bir embedding layer olmal
                                         # modelim ka token ile eitelecek, yukarda hatrlarsanz 15000 vermitik. 
                                         # Test setimde 15000 haricinde bir token kullanlrsa test setimdeki token ignore edilecek
                    output_dim=embedding_size, # Sana gnderdiim her dizin 61 tokenden oluur. 
                                               # Bunlar 15000 token ile eitilecek. 
                                               # Embedding layer dan kalk vektre dnn                    
                    input_length=max_tokens)) # Benim modele 100 lk vektrler halinde girecek.

model.add(Dropout(0.2))

model.add(GRU(units=48, return_sequences=True))  # units : nron ays ( embedding size / 2)
                                                 # nron saylar 2 ve zeri saylar belirlemek best practise dir. 
                                                 # Direkt 500 veya 100 de diyebilirdik
                                                 # "return_sequences=True" GRU modelleri dizinlere aktarrken true yaparz. 
                                                 # bir sonraki katman LSTM veya GRU ise kullanlr.
                                                 # 'return_sequences=True' ile cell state ve hidden state dngsn salar.
model.add(Dropout(0.2))

#model.add(GRU(units=24, return_sequences=True))  
#model.add(Dropout(0.2))

model.add(GRU(units=12)) # Bir sonraki katman GRU veya LSTM olmad iin "return_sequences=True" kullanmadk.

model.add(Dense(1, activation='sigmoid')) # 0-1 lerden oluan output almak iin SIGMOID


# modelime verdiim dizinlerin iinde 61 token var
# her tokeni belirlediimiz 61 saysna gre boyutlandryor
# dizin iindeki 61 tokeni 100 lk vektrlere evir

# input_dim : modelim ka token ile eitilecek (15000)
# output_dim : 15000 token gelsin bunlarn hepsini 100 lk vektrlere dntr ve activasyon function lara gnder
# input_length : sana gnderdiim her dizin 61 token den oluuyor. 

# 15000 token olmal ve her bir token 100 lk vektrlerden olumal
```

```python
# output_dim : 15000 token gelsin bunlarn hepsini 100 lk vektrlere dntr ve acrivasyon function lara gnder
# input_dim : modelim ka token ile eitilecek (15000)
# 61 den 100 boyuta m kyor embedding ile
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(monitor="val_recall", mode="max", 
                           verbose=1, patience = 1, restore_best_weights=True)

# restore_best_weights : 
    # arlklar en iyi scora gncelle. 
    # en iyi skoru yakalamak iin kullanlr. 
    # yoksa patient sonundaki skoru verir.
    # nk patience 5 dediimiz iin en kt skordan sonra 5 skor daha alyor ve ald son skoru getiriyor
```

```python
pd.Series(y_train).value_counts(normalize=True)  # class larn oranna bakar ve arlklarn dengeleriz.
```

```python
from sklearn.utils import class_weight
classes_weights = class_weight.compute_sample_weight(class_weight='balanced', y=y_train)
pd.Series(classes_weights).unique()
```

```python
# Bu ekilde manuel de verebiliriz.

# weights = {0:95, 1:5}

# 0,94  100 (5) e tamamlayacak ve 0,05 i 100 (95) e tamamlayacak ekilde arlk veriyoruz. 
# dorudan int say kullanyoruz. 
# arlklar bu ekilde belirleniyor.
```

```python
model.fit(X_train, y_train, epochs=10, batch_size=256, sample_weight= classes_weights,
         validation_data=(X_test, y_test), callbacks=[early_stop])
```
