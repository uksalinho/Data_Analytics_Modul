**Adım -1 :** Küme sayısını tespit edin (k)

**Adım - 2:** Rastgele gözlemler seçin ve bu gözlemleri ilk merkezler yapın.

**Adım - 3:** İlk merkezlere olan mesafeler her gözlem için hesaplanır.

**Adım -5 :** Atama sonrası oluşan kümeler için tekrar merkez yapılır.

**Adım -6 :** Yinelemelerden sonra, küme içi hata kareleri toplamının minimum olduğu durumda gözlemlerin küme yapısı son küme olarak seçilir.

**[DE]**

**Schritt -1:** Ermitteln Sie die Anzahl der Cluster (k)

**Schritt - 2:** Wählen Sie zufällige Beobachtungen aus und machen Sie diese Beobachtungen zu ersten Schwerpunkten.

**Schritt - 3:** Entfernungen zu den ersten Schwerpunkten werden für jede Beobachtung berechnet.

**Schritt - 4:** Weisen Sie jeden Datenpunkt dem nächsten Schwerpunkt zu.

**Schritt -5:** Schwerpunkte werden erneut für die nach der Zuweisung gebildeten Cluster erstellt.

**Schritt –6:** Nach den Iterationen wird die Clusterstruktur der Beobachtungen in dem Fall, in dem die Summe der Intra-Cluster-Fehlerquadrate minimal ist, als endgültiger Cluster ausgewählt.

**[TR]**

**Hopkins testi**, bir değişkenin rastgeleliği için istatistiksel bir test.

**Boş Hipotez (H0) :** Veri noktaları rastgele olmayan, tek tip dağılımla oluşturulur (anlamlı kümeler anlamına gelmez)

**Alternatif Hipotez (Ha):** Veri noktaları rastgele veri noktaları tarafından üretilir(kümelerin varlığı)

**[DE]**

**Hopkins-Test**, ein statistischer Test für die Zufälligkeit einer Variablen.

**Nullhypothese (H0):** Datenpunkte werden durch nicht zufällige, gleichmäßige Verteilung generiert (was bedeutet, dass keine sinnvollen Cluster vorhanden sind)

**Alternative Hypothese (Ha):** Datenpunkte werden durch zufällige Datenpunkte generiert (Vorhandensein von Clustern)

**[TR]**

**Düzeltilmiş Rand Endeksi**

- Düzeltilmiş Rand İndeksi, tüm örnek çiftlerini göz önünde bulundurarak ve tahmin edilen ve gerçek kümelemede aynı veya farklı kümelere atanan çiftleri sayarak iki kümeleme arasındaki benzerlik ölçüsünü hesaplar.

**[DE]**

**Angepasster Rand-Index**

- Der Adjusted Rand Index berechnet ein Ähnlichkeitsmaß zwischen zwei Clustern, indem er alle Musterpaare berücksichtigt und Paare zählt, die im vorhergesagten und im wahren Cluster demselben oder unterschiedlichen Clustern zugeordnet sind.

- Der Wert von ARI zeigt kein gutes Clustering an, wenn er nahe Null oder negativ ist, und einen guten Cluster, wenn er nahe 1 ist.

**[TR]**

**Siluet Katsayısı**

- Doğruluk etiketleri bilinmiyorsa, modelin kendisi kullanılarak değerlendirme yapılmalıdır.
(Değerlendirme yöntemlerinden biri Silhouette Coeffícient'dir)

- Daha yüksek bir Siluet Katsayısı puanı, daha iyi tanımlanmış kümelere sahip bir modelle ilgilidir.

                - a : Bir numune ile aynı sınıftaki diğer tüm noktalar arasındaki ortalama mesafe.
                - b : Bir örnek ile bir sonraki en yakın kümedeki diğer tüm noktalar arasındaki ortalama mesafe.

- Tek bir numune için Siluet Katsayısı s şu şekilde verilir:
                             
                             a = (b - a ) / maks(a,b)

                
                -1 (yanlış kümeleme) <-- S --> +1(çok yoğun kümeleme)


**[DE]**

**Silhouettenkoeffizient**

- Wenn die Ground-Truth-Labels nicht bekannt sind, muss die Auswertung mit dem Modell selbst durchgeführt werden.
(Eine der Bewertungsmethoden ist Silhouette Coeffícient)

- Ein höherer Silhouettenkoeffizient bezieht sich auf ein Modell mit besser definierten Clustern.

                - a : Der mittlere Abstand zwischen einer Probe und allen anderen Punkten in derselben Klasse.
                - b : Der mittlere Abstand zwischen einer Probe und allen anderen Punkten im nächstnächsten Cluster.

- Der Silhouetten-Koeffizient s für eine einzelne Probe wird dann wie folgt angegeben:
                             
                             a = (b - a ) / max(a,b)


                  -1 (falsches Clustering) <-- S --> +1 (sehr dichtes Clustering)

# Choosing The Optimal Number of Clusters

## Elbow metod

```python
# k sayisi ne olcak ?
# ssd : sum of squared distance
# inertia : her bir gözlemin centroid e uazakliginin karesi alinip toplandiginda ne kadar minimum deger aliyorsa 
# kendi icinde ne kadar siki sikiya kümelendigini göstermektedir.

# Orion hoca --> inertia : Sum within Cluster Distance
```

```python
ssd = []

K = range(2,10)

for k in K:
    model = KMeans(n_clusters =k, random_state=42)
    model.fit(X)
    ssd.append(model.inertia_)
    
# k = 1 olmasi demek sadece bir kümeden olusmasi demek ve cok hata olusuyor.
```

```python
plt.plot(K, ssd, "bo-")
plt.xlabel("Different k values")
plt.ylabel("inertia-error") 
plt.title("elbow method") 

# keskin düsüsün durdugu ilk nokta 3 tür. Her zaman ilk noktayi aliyoruz.
```

```python
ssd # sum of squared distance
```

```python
pd.Series(ssd).diff() #diff bir önceki gözlem ile kendi arasindaki farki aliyor.
```

```python
df_diff =pd.DataFrame(-pd.Series(ssd).diff()).rename(index = lambda x : x+1)
df_diff
```

```python
df_diff.plot(kind='bar') # keskin düsüsün durdugu ilk nokta, 3 noktasi oldugu asikar
```

```python
range_n_clusters = range(2,9)
for num_clusters in range_n_clusters:
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(X)
    cluster_labels = kmeans.labels_
    # silhouette score
    silhouette_avg = silhouette_score(X, cluster_labels)
    print(f"For n_clusters={num_clusters}, the silhouette score is {silhouette_avg}")
    
    # n_cluster = 2 de en iyi skoru da alsak;
    # Burada Elbow method ve görsellere de bakip 3 e karar veriyoruz. Daha iyi kümelenmenin insight ini veriyor.
```

## Silhouette analysis

```python
range_n_clusters = range(2,9)
for num_clusters in range_n_clusters:
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(X2)
    cluster_labels = kmeans.labels_
    # silhouette score
    silhouette_avg = silhouette_score(X2, cluster_labels)
    print(f"For n_clusters={num_clusters}, the silhouette score is {silhouette_avg}")
```
