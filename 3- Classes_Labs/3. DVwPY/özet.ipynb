pairplot

___________
sns.pairplot(tips, hue='sex', palette="coolwarm");  # palette ile hue parametresine renklendirme yapilabilir
___________
sns.pairplot(tips, hue='sex', palette='viridis', corner=True);  # corner=True ile üstteki degerler olmadan bastiirilabilir
___________
g = sns.PairGrid(tips, hue="sex", palette="viridis")

g.map_upper(sns.scatterplot)
g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot);  # grafigin neresine ne istiyorsam
___________
g = sns.PairGrid(tips)

g.map_upper(sns.scatterplot, color="purple")
g.map_diag(sns.histplot, color="green")
g.map_lower(sns.kdeplot, color="red");  # grafigin neresine hangi renk istiyorsam
___________
sns.pairplot(tips, hue='sex', palette='viridis', corner=True, diag_kind="Auto");  # diag_kind="Auto" kind ile diagnogal'i bosaltti. diag_kind: {‘auto’, ‘hist’, ‘kde’, None}. # diag_kind yazmayinca default kde/hist gelir

___________

clustermap

sns.clustermap(tips.corr());  # seaborn'un clustermap() işlevi, verilen matris veri kümesinin hiyerarşik olarak kümelenmiş bir                                 ısı haritasını çizer. Kümelenmiş bir ızgara dizini döndürür.
clustermap() yöntemi, verileri benzerliğe göre sıralamak için hiyerarşik(hierarchically) kümeler kullanır. Bu, satırlar ve sütunlar için verileri yeniden düzenler ve verileri daha da derinlemesine anlamak için benzer içeriği yan yana görüntüler.
