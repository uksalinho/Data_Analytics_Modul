## Word2Vec

```python
# pip install gensim
```

```python
import warnings
warnings.filterwarnings("ignore")

from gensim.models import Word2Vec
```

```python
# zipli haliyle kullanmak icin
# import zipfile

# text = zipfile.ZipFile('newspaper.zip')
# text = text.read('newspaper.txt')
# text = text.decode('utf-8')

# text[:1000]
```

```python
#zip'den cikmis haliyle
text = open('newspaper.txt', 'r', encoding='utf8')  # türkce karakterleri okuyabilmek icin "utf8"
text = text.read()
text[:1000]  # ilk 1000 karateri aldim
```

```python
list_sent = text.split('\n')  # split('\n') ile liste icinde cümlelere ayir
list_sent[:10]  # ilk 10 cümleyi göster
```

```python
list_sent[0].split()  # 0. cümleyi word tokenlerine ayirdim
```

```python
corpus = []

for sent in list_sent:  # her bir cümleyi liste icinde tokenlerine ayirdi
    corpus.append(sent.split())
```

```python
print(corpus[:10])  # yine ilk 10a baktim
```

```python
model = Word2Vec(corpus, vector_size=100, window=5, min_count=5, sg=1)

# window=5 genelde 5 ya da 10 secilir
# window=5 tokenin 5 önce ve 5 sonraki tokenlarla iliskisini dikkate alir
# vector_size=100 : vektörümüzün kaç boyutlu olmasını istiyorsak. (bazı sürümlerde "size" olarak da geçiyor)
# min_count=5: bir token 5 ten fazla ise eğitime dahil et. az ise eğitime dahil etme
# skip-gram küçük corpuslarda iyi sonuç verirken, CBOW çok daha büyük corpuslarda iyi sonuç veriyor.
# default sg=0 bu durumda CBOW kullaniyor
```

```python
model.wv['ankara']

# ankara için bana 100 boyutlu bir vektör verdi.
# bu vektör 100 tane farklı representation ı temsil ediyor.
```

```python
model.wv.most_similar('öğretmen')

# ögretmen tokeni ile en fazla iliskili tokenleri buluyor
# önce kendisi yerine ikame kelimeler ve sonrasında birşey ile beraber kullanılanlar (anlamsal ilişkiler atamış)
```

```python
model.wv.most_similar('kırmızı')

# eğittiğim corpus a göre, "birşey" kelimesi ile anlamsal yakınlığa sahip olan kelimeleri buluyor. 
  # yanındaki değerler de ne kadar ilişkili olduğunu gösteriyor.. 

# Similar; anlamsal ilişkiyi bulurken yan yana geçen kelimeleri  ve yerine ikame edilen kelimeleri ayrı ayrı değerlendiriyor.
    # bunu yaparken önce yerine ikame edilen kelimelere bakıyor. onu similar olarak kaydediyor. 
        # eğer ikame kelime bulamazsa yan yana geçen kelimelere bakarak benzerlik ağırlığını belirliyor
```

```python
model.wv.most_similar('eve')

# "eve" ile birlikte kullanılan kelimeler ile hangi başka kelimeler birlikte kullanılıyor ise onlar eve nin yerine ikame edilen kelimelerdir.
  # mesela eve gittim i ele alıp, gittim le birlikte kullanılan diğer kelimeleri aşağıda sıraladığını görüyoruz. 
   # (dükkana gittim, ormana gittim, mağazaya gittim gibi) 
```

```python
model.wv.most_similar('mavi')

 # buradaki skorlar 0-1 arasında. cos benzerliği
```

```python
model.wv.most_similar(positive=['öğrenme', 'doktor'], negative=['tedavi'], topn=5)
```

```python
model.wv.most_similar(positive=['ankara', 'belçika'], negative=['brüksel'], topn=1)
```

```python
model.save("word2vec.model")

# modeli kaydetme
```

## Glove

```python
from gensim.models import KeyedVectors
```
