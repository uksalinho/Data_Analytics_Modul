## Cleaning Data

```python
import re
s = "http\\:www.mynet.com #lateflight @airlines"
s = re.sub("http\S+", "", s).strip()  # http sonrasi bosluk haric en az bir karakter
s

# re (regex) fonksiyonu ile .sub metodunu kullanarak s textinin icerisindeki gereksiz stringleri replace yapabiliriz.
# strip() ile bosluklari siliyoruz.
```

```python
s = re.sub("#\S+", "", s)
s
```

```python
s = re.sub("@\S+", "", s)
s.strip()
```

```python
# \w typically matches [A-Za-z0-9_]

# str.fonksiyonlarn (pandastaki fonksiyonlar) uygulayabilmek iin ncelikle Series haline getirmek gerekli.
# Aksi takdirde hata verir
```

```python
stop_words = stopwords.words('english')

for i in ["not", "no"]:
        stop_words.remove(i)
        
# olumsuz yorumlari da tesbit edebilmak icin not ve no ifadelerini stopwords listesinden cikardik
```

```python
def cleaning(data):
    
    import re
    # import islemini def in icerisinde yapmak gerekiyor
    
    #1. Removing URLS
    data = re.sub('http\S+', '', data).strip()
    data = re.sub('www\S+', '', data).strip()

    #2. Removing Tags
    data = re.sub('#\S+', '', data).strip()

    #3. Removing Mentions
    data = re.sub('@\S+', '', data).strip()
    
    #4. Removing upper brackets to keep negative auxiliary verbs in text
    data = data.replace("'", "")
    
    # ayralar kaldryorum ki (can't gibi) olumsuzluk ieren kelimeler iinde ayra ierdii iin stopwords lerde silinmesin.
    # Yorumun olumsuz haliyle de olsa kalmas iin ilave bir yntem olarak kullanld.
    
    #5. Tokenize
    text_tokens = word_tokenize(data.lower())
    
    #6. Remove Puncs and number
    tokens_without_punc = [w for w in text_tokens if w.isalpha()]
    
    # punctuation ve no lardan kurtuluyoruz.
    
    #7. Removing Stopwords
    tokens_without_sw = [t for t in tokens_without_punc if t not in stop_words]
    
    # stopword lerden temizleme
    # tokenleri teker teker al stopwords ler iinde yoksa olduu gibi yaz, yani stopwords ise yazma.
    # storwards olmayanlar liste iine yazmaya devam et   
    
    #8. lemma
    text_cleaned = [WordNetLemmatizer().lemmatize(t) for t in tokens_without_sw]
    
    # stopwardslardan temizlenmi tokenleri alyoruz ve lemma uyguluyoryuz
    
    #joining
    return " ".join(text_cleaned)

     # tamamen temizlenmi ve kklerine inilmi kelimeleri alp join yapyoruz.
```

```python
# def aklamas:
# - iki tane def te ayn, aadaki bir nceki derste gsterilendir.
# - yukardaki def fonksiyonunda ayralar kaldrma ilemi gerekletirdik. upgrade ettik.
# - 1. ayralar kaldrld ve kk harfe dntrld.
# - 2. noktalama iaretler ve numaralardan arndrlma ilemi yapld
# - 3. token without punc iindeki tm tokenleri al ve iinden stop words olmayanlar ile yola devam et.
# - 4. stop word lerden arndrlm tokenlerin lemma yani kklerine gittik.
# - 5. text imizi birletirdik.
```

```python
cleaning_text = df2["text"].apply(cleaning)
cleaning_text.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df2["text"]
y= df2["airline_sentiment"]

# temizleme ilemi bittii iin artk data setimi LABEL ve FEATURE olarak 2 ye ayrabiliriz.
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=101)

# veri dengesiz-imbalanced olduu iin stratify kullanld.
```

## Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# max_df is used for removing terms that appear too frequently, also known as "corpus-specific stop words". For example:

# max_df = 0.50 means "ignore terms that appear in more than 50% of the documents".
# max_df = 25 means "ignore terms that appear in more than 25 documents".
# The default max_df is 1.0, which means "ignore terms that appear in more than 100% of the documents". Thus, the default setting does not ignore any terms.

# min_df is used for removing terms that appear too infrequently. For example:

# min_df = 0.01 means "ignore terms that appear in less than 1% of the documents".
# min_df = 5 means "ignore terms that appear in less than 5 documents".
# The default min_df is 1, which means "ignore terms that appear in less than 1 document". Thus, the default setting does not ignore any terms.
```

```python
vectorizer = CountVectorizer(preprocessor=cleaning, min_df=3) # ngram_range=(1,2), max_features= 1500

# benim corpusum (datasetim) ierisindeki geen btn unique tokenlerimi nce tespit eden ve 
# ardndan sayan bir ilemdir.

X_train_count = vectorizer.fit_transform(X_train)

# fit ile X_trainde geen unique tm tokenleri tespit ediyor ve transform ile onlar sayyor

X_test_count = vectorizer.transform(X_test)

# X_train de eitildii iin X_train de geme durumuna gre X_test'te sayma ilemini yapar.
# X_testte geen bir token, eitildii X_train de yok ise onu yok sayar, ilem yapmaz

# min_df = 3    --> corpusumda 3'den daha az document'da (yorumda) geen tokenleri ignore et.
# max_df = 1000 --> corpusumda 1000'den daha fazla document'da (yorumda) geen tokenleri ignore et.

# max_features = 1500 --> corpusumda en sk kullanlan ilk 1500 feature (tokeni) dikkate al dierlerini ignore et.


# Egitime katkisi olmayacak tokenleri almamak icin min_df=3 aldik, 
# yani token sayisi 3 ve daha az olanlari degerlendirmeye almiyoruz.
# max_df=0.5 tm dkmanlarin %50 sinde geciyorsa ignore et anlamina geliyor. Cok sik tekrar eden token lari elemis oluyoruz.

# ngram_range kelimeleri gruplayip token olarak vektrize islemi uyguluyor. rnegin;
# ngram_range=(1,3) alindiginda ardarda gelen 1'li, 2'li ve 3'l kelime gruplarini token yaparak sayiyor.
# Birlikte kullanildiginda farkli bir anlam cikan, kelime tamlamalari ya da kaliplarini yakalamak amaci ile ngram_range kullanilir.

# max_features=n ile tokenler icerisinden en fazla kullanilan 1500 tanesini belirleyip vektrize islemi yapiyor.

# demek ki X_train yeterince byk olmal ki X_test iinde geen her trl olasla cevap versin, tm tokenler iin sayma ilemini yapsn, kelime atlamasn. 

# scale ileminde yaptmz gibi, leakage nlemek iin train --> fit_transform, test --> transform
# fit_transform dediimiz iin unique olan tm kelimeleri sayyor ve array e dntryor
# test te geen bir kelime train de gemiyorsa model o kelimeyi yok sayar, grmez.
# model ne kadar fazla kelime ieriyorsa o kadar bu handikapn nne geiyor.

# FIT train setindeki kelimeleri tespit ediyor, transform sayyor ve array a dntryor
# FIT train setine gre yapld iin, TEST te TRANe gre transform olacak.
```

## Model Comparisons - Vectorization

```python
from sklearn.metrics import confusion_matrix, classification_report, f1_score, recall_score
```

```python
def eval(model, X_train, X_test):
    y_pred = model.predict(X_test)
    y_pred_train = model.predict(X_train)
    print(confusion_matrix(y_test, y_pred))
    print("Test_Set")
    print(classification_report(y_test,y_pred))
    print("Train_Set")
    print(classification_report(y_train,y_pred_train))
    
```

## Naive Bayes

```python
# A olay gerekletiinde B olaynn gerekleme olasl nedir? (P(A|B))
# B olay gerekletiinde A olaynn gerekleme olasl nedir? (P(B|A))
# Datasetime nasl uygulayabilirim?
# P(dont | negative) --> dont yardmc fiili getiinde bunun negatif olma ihtimali nedir
# P(dont enjoy|positive) --> dont ve enjoy tokenleri birlikte getiinde positive olma ihtimali nedir
```

```python
from sklearn.naive_bayes import MultinomialNB, BernoulliNB # BernoulliNB for binary model

# BernoulliNB for binary model
# MultinomialNB, modelimizde birden fazla label var ise bunun kullanlmas tavsiye ediliyor.

# dkmannda multinomial ve bernoille den ikisini de deneyip hangisi daha iyi sonu veriyorsa onunla devam edilmesi tavsiye ediliyor.
# (nceden denenmi ve MulitnomialNB daha iyi km.)
```

```python
nb = MultinomialNB(alpha=10)
nb.fit(X_train_count,y_train)

# burada artk saysallatrlm vektrleri kullanyoruz. text olarak hibir ey brakmyoruz.
# Alpha arttikca overfittingi engelliyor. Regularization icin alpha katsayisi kullaniliyor.
```

```python
# Aada her model iin aldmz ilkel skorlar CROSS VALIDATe yaparak uygun olup olmadklarn, uyuup uyumadklarn grebiliriz.
# Bu almada ilemler uzun srmesin diye modeller iin ayr ayr CV almas yaplmad
# Sadece seilen final modelde yapld
```

```python
print("NB MODEL")
eval(nb, X_train_count, X_test_count)

# skorlarm dengeli. 
# dolaysyla overfitting olmadn anlyoruz.
# negatifleri iyi yakalam
# az olmasna ramen pozitifleri nispeten iyi yakalam
```

```python
from sklearn.model_selection import cross_validate
from sklearn.metrics import make_scorer
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score

scoring = {'accuracy': make_scorer(accuracy_score),
            'precision-neg': make_scorer(precision_score, average=None, labels=["negative"]),
            'recall-neg': make_scorer(recall_score, average=None, labels = ["negative"]),
            'f1-neg': make_scorer(f1_score, average=None, labels = ["negative"])}

model = MultinomialNB(alpha=10)
scores = cross_validate(model, X_train_count, y_train, scoring = scoring, cv = 10, return_train_score=True)
df_scores = pd.DataFrame(scores, index = range(1, 11))
df_scores.mean()[2:]
```

```python
from yellowbrick.classifier import PrecisionRecallCurve

viz = PrecisionRecallCurve(
    MultinomialNB(alpha=10),
    classes=nb.classes_,   # label larn isimleri ile gzkmesi iin.. aksi halde isim yerine 0,1,2 diye gzkr 
    per_class=True,
    cmap="Set1"
)
viz.fit(X_train_count,y_train)
viz.score(X_test_count, y_test)
viz.show();

# datada dengesizlik (umbalance) sz konusu olduu iin skorlarm gereki olsun diye viz = PrecisionRecallCurve u kullandm.
# data imbalance olduu iin precision recall kullanld
```

## TF-IDF

```python
# scale ileminde yaptmz gibi, leakage nlemek iin train --> fit_transform, test --> transform


# fit_transform dediimiz iin unique olan tm kelimeleri sayyor, hangi satrlarda getiini buluyor ve array e dntryor
# test te geen bir kelime train de gemiyorsa model o kelimeyi yok sayar, grmez.
# model ne kadar fazla kelime ieriyorsa o kadar bu handikapn nne geiyor.

# TF
# FIT train setindeki unique tokenleri tespit ediyor ve hangi satrlarda getiini buluyor, 
# transform bu tokenleri ve getii satrlar sayyor
# FT train setine gre yapld iin, TEST te TRANe gre transform olacak.
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# eitim mant dierleriyle tamamen ayn

tf_idf_vectorizer = TfidfVectorizer(preprocessor=cleaning, min_df=3)
X_train_tf_idf = tf_idf_vectorizer.fit_transform(X_train)   # FIT train setindeki unique tokenleri tespit ediyor ve hangi satrlarda getiini buluyor
X_test_tf_idf = tf_idf_vectorizer.transform(X_test)    # TRANSFORM bu tokenlerin saysn alyor ve bu tokenlerle TF-IDF yapyor

# tf_idf: fit deyince X_traindeki tm unique tokenleri tespit ediyor ve btn satrlarda (dokmanlarda) o tokenin geme skln da sayyor.
# count vectorizer da sadece unique tokenleri tespit ediyordu burda dokmanlardaki tekrar says da iin iinde..

# transform deyince bunlar sayyor ve fonksiyonu uygulayp tf-idf katsaylarn buluyor.
```

```python
# TF = KELMEM / TOPLAM KELME SAYISI
# DF = CORPUS TAK TOPLAM DOKMAN SAYISI / KELMEMN GET DOKMAN SAYISI
```

```python
# TF(t) = (Number of times term t appears in a document) / (Total number of terms in the document)

# IDF(t) = log(Total number of documents  in a corpus / Number of documents with term t in it)
```

```python
# Example: 
# Consider a document containing 100 words wherein the word cow appears 3 times
# The TF for cow is then (3 / 100) = 0.03 
# Now, let's assume the corpus has 10 million documents and the word cow appears in one thousand of these
# Then, the IDF is calculated as log(10,000,000 / 1,000) = 4
# Thus, the Tf-idf weight is the product of these quantities: 0.03 * 4 = 0.12
```
