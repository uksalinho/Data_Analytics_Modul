```python
benzine =['Gasoline', 'Super 95', 'Regular', 'Super E10 95', 'Super Plus 98', 'Super Plus E10 98', 'Others']
lpg = ['LPG', 'Liquid petroleum gas', 'CNG', 'Biogas', 'Domestic gas H']

def fueltype(x):
    if x in benzine:
        return 'Benzine'
    elif x in lpg:
        return 'LPG/CNG'
    else:
        return x
df['Fuel'] = df.Fuel.apply(fueltype)
```
